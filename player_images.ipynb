```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import asyncio
import json
import aiohttp
from understat import Understat
import nest_asyncio
```

```python
players = pd.read_csv('player_data_full.csv',low_memory=False)
players = players[["name", "full_name", "short_name", "image", "club_name", "club_logo", "country_name", "country_flag", "club_kit_number"]]
```

```python
players.head()
```

```python
nest_asyncio.apply()
async def teams():
    async with aiohttp.ClientSession() as session:
        understat = Understat(session)
        teams = await understat.get_teams(
            "epl",
            2024
        )
        return(json.dumps(teams))

teams = asyncio.run(teams())
teams = json.loads(teams)
```

```python
team_list = []

for x in teams:
    team_list.append(x["title"])
```

```python
def in_league(row):
    if row["club_name"] in team_list:
        row["league"] = "Premier League"
    else:
        row["league"] = "Other"
    return row

players = players.apply(lambda x: in_league(x), axis=1)
```

```python
players = players[players["league"] == "Premier League"]
```
